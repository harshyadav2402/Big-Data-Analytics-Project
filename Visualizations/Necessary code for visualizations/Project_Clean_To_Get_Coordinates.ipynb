```python
import googlemaps
from datetime import datetime
import pandas as pd

gmaps = googlemaps.Client(key='AIzaSyBBejjT6RgpCU9jGVjefT0oVeONqcbvI9g')
df = pd.read_csv('10003_Fire_Incident_Dispatch_Data_2016.csv')

#adding latitude and longitude columns to the imported dataframe
df['LATITUDE'] = 0.0
df['LONGITUDE'] = 0.0
```

```python
no_of_unknown_addresses = 0
no_of_confirmed_addresses_within_zipcode = 0
#iterating over the dataframe to extract latitude and longitude of every fire incident
for i in range(len(df)):
    try:
        geocode_result = gmaps.geocode(df.iloc[i]['ALARM_BOX_LOCATION'])
        x = len(geocode_result[0]['address_components'])
        if (int(geocode_result[0]['address_components'][x-1]['long_name'] )== df['ZIPCODE'].iloc[i]):
            df['LATITUDE'].iloc[i]= geocode_result[0]['geometry']['location']['lat']
            df['LONGITUDE'].iloc[i] = geocode_result[0]['geometry']['location']['lng']
            no_of_confirmed_addresses_within_zipcode = no_of_confirmed_addresses_within_zipcode + 1
            print(i)
            
    except:
        no_of_unknown_addresses = no_of_unknown_addresses + 1
        
print("number of unknown addresses: ", no_of_unknown_addresses)
print("no of confirmed addresses within zipcode: ", no_of_confirmed_addresses_within_zipcode)
```
